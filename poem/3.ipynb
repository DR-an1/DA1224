```python
#柱状图
import pandas as pd
from pyecharts import Bar,Line,Overlap


data=pd.read_excel('c:/daima/dataanalysis/poem/100.xlsx')
data=data.groupby(['dynasty']).sum()
attr=['两汉','五代','元代','先秦','南北朝','唐代','宋代','明代','清代','近代','魏晋']
v1=data['goods']/10000
v2=data['counts']
bar = Bar("历代诗歌数据对比")
bar2 = Bar()
bar2.add("数量", attr, v2,is_label_show=True)
bar.add("点赞数(万)", attr, v1,is_label_show=True)
overlap = Overlap()
# 默认不新增 x y 轴，并且 x y 轴的索引都为 0
overlap.add(bar)
# 新增一个 y 轴，此时 y 轴的数量为 2，第二个 y 轴的索引为 1（索引从 0 开始），所以设置 yaxis_index = 1
# 由于使用的是同一个 x 轴，所以 x 轴部分不用做出改变
overlap.add(bar2, yaxis_index=1, is_add_yaxis=True)
overlap.render()
overlap
```
