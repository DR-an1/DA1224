```python
#前100
import pandas as pd
from pyecharts import Scatter


ds=pd.read_excel('C:/daima/dataanalysis/poem/诗.xlsx')
df=pd.read_excel('C:/daima/dataanalysis/poem/词.xlsx')
dq=pd.read_excel('C:/daima/dataanalysis/poem/曲.xlsx')
dw=pd.read_excel('C:/daima/dataanalysis/poem/文.xlsx')
ds=ds.drop_duplicates()
df=df.drop_duplicates()
dq=dq.drop_duplicates()
dw=dw.drop_duplicates()
ds["good"]=ds.good.apply(lambda x:int(x))
df["good"]=df.good.apply(lambda x:int(x))
dq["good"]=dq.good.apply(lambda x:int(x))
dw["good"]=dw.good.apply(lambda x:int(x))
all=pd.concat([ds,df,dq,dw])
grouped = all.groupby(['name','dynasty'])
grouped_pct = grouped['good']
new=grouped_pct.agg(['sum','count'])
news=new.sort_values('sum',ascending=False)[0:100]
'''
v1=news['count']
v2=news['sum']
scatter = Scatter("散点图示例")
scatter.add('a', v1, v2)
scatter.render()
scatter
'''
news
```
