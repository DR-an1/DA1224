```python
#箱型图(工资与是否大公司的关系)
import os
import pandas as pd
from pyecharts import Bar,Boxplot
os.chdir('C:/爬虫/拉勾网')

data=pd.read_excel('C:/爬虫/拉勾网/all.xlsx')
grouped=data.groupby(['经验'])
grouped_pct=grouped['工资']
aprice=grouped_pct.agg(['mean','count'])

'''
y1=data['aSalary'][data['Size']==' 少于15人']
y2=data['aSalary'][data['Size']==' 15-50人']
y3=data['aSalary'][data['Size']==' 50-150人']
y4=data['aSalary'][data['Size']==' 150-500人']
y5=data['aSalary'][data['Size']==' 500-2000人']
y6=data['aSalary'][data['Size']==' 2000人以上']
y6
boxplot = Boxplot("工资与公司规模透视")
x_axis = [ '少于15人','15-50人', '50-150人', '150-500人','500-2000人','2000人以上']
y_axis = [
    y1,
    y2,
    y3,
    y4,
    y5,
    y6
]
_yaxis = boxplot.prepare_data(y_axis)       # 转换数据
boxplot.add("月薪(元)", x_axis, _yaxis,xaxis_rotate=30)
boxplot.render()
boxplot
'''
#平均薪资
aprice['mean']=round(aprice['mean'],0)
attr = ['经验1-3年', '经验10年以上', '经验1年以下', '经验3-5年', '经验5-10年', '经验不限', '经验应届毕业生']
v1 = aprice['mean']
bar = Bar("深圳IT从业者平均工资")
bar.use_theme("macarons")
bar.add("薪资(元)", attr, v1, is_stack=True,  is_label_show=True,xaxis_rotate=30)
bar.render()
bar
```
