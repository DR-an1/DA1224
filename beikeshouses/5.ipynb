```python
import os
import pandas as pd
from pyecharts import Bar,Boxplot
os.chdir('C:/爬虫/拉勾网')

data=pd.read_excel('C:/爬虫/拉勾网/all.xlsx')
grouped=data.groupby(['学历'])
grouped_pct=grouped['aSalary']
aprice=grouped_pct.agg(['mean','count'])
aprice
y1=data['aSalary'][data['学历']==' 本科']
y2=data['aSalary'][data['学历']==' 不限']
y3=data['aSalary'][data['学历']==' 大专']
y4=data['aSalary'][data['学历']==' 硕士']

boxplot = Boxplot("工资与学历关系透视")
x_axis = [ '本科','不限', '大专', '硕士']
y_axis = [
    y1,
    y2,
    y3,
    y4,
]
#_yaxis = boxplot.prepare_data(y_axis)       # 转换数据
boxplot.add("月薪(元)", x_axis, y_axis,xaxis_rotate=30)
boxplot.render()
boxplot
```
