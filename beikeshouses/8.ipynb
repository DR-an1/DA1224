```python
#整租房(平均价)
import os
import pandas as pd
from pyecharts import Bar,Boxplot
os.chdir('C:/爬虫/深圳租房')

data=pd.read_excel('C:/爬虫/深圳租房/深圳整租房信息.xlsx')
nanshan=data[data.地区=='南山区'].mean()
nanshan
grouped=data.groupby(['地区'])
grouped_pct=grouped['价格']
aprice=grouped_pct.agg(['mean','count'])
aprice['mean']=round(aprice['mean'],0)
aprice

#整租和合租数量
attr = ['南山区', '宝安区', '福田区', '罗湖区', '龙华区', '龙岗区','光明新区', '大鹏新区', '坪山区','盐田区']
v1 = [570,226,747,166,311,570,0,0,0,0]
v2 = [1300,821,1265,804,789,1311,37,9,10,96]
bar = Bar("")
bar.use_theme("macarons")
bar.add("合租", attr, v1, is_label_show=True,xaxis_rotate=30)
bar.add("整租", attr, v2, is_label_show=True,xaxis_rotate=30)
bar.render()
bar
```
