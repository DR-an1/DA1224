```python
#各语言箱型图
import os
import pandas as pd
from pyecharts import TreeMap,Bar,Boxplot

data1=pd.read_excel('C:/爬虫/拉勾网/Python岗位.xlsx')
data2=pd.read_excel('C:/爬虫/拉勾网/Java岗位.xlsx')
data3=pd.read_excel('C:/爬虫/拉勾网/C++岗位.xlsx')
data4=pd.read_excel('C:/爬虫/拉勾网/C岗位.xlsx')
data5=pd.read_excel('C:/爬虫/拉勾网/php岗位.xlsx')
data6=pd.read_excel('C:/爬虫/拉勾网/.net岗位.xlsx')

boxplot = Boxplot("")
x_axis = ['Python', 'Java', 'C++', 'C', 'php', 'C#']
y_axis = [
    data1['aSalary'],
    data2['aSalary'],
    data3['aSalary'],
    data4['aSalary'],
    data5['aSalary'],
    data6['aSalary'],
]
_yaxis = boxplot.prepare_data(y_axis)       # 转换数据
boxplot.add("boxplot", x_axis, _yaxis)
boxplot.render()
boxplot

#人数
attr = ['Python', 'Java', 'C++', 'C', 'php', 'C#']
v1 = [data1['Name'].count(), data2['Name'].count(), data3['Name'].count(), data4['Name'].count(), data5['Name'].count(), data6['Name'].count()]
bar = Bar("")
bar.use_theme("macarons")
bar.add("", attr, v1, is_stack=True,  is_label_show=True)
bar.render()
bar
```
