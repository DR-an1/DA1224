```python
import os
import pandas as pd
from pyecharts import Bar
os.chdir('C:/爬虫/拉勾网')

data=pd.read_excel('C:/爬虫/拉勾网/数据分析岗位finals.xlsx')
grouped=data.groupby(['Area'])
grouped_pct=grouped['aSalary']
aprice=grouped_pct.agg(['mean','count'])
aprice
aprice['mean']=round(aprice['mean'],0)
attr = ["光明新区", "南山区", "宝安区", "盐田区","福田区", "罗湖区", "龙华新区","龙岗区"]
v1 = aprice['mean']
bar = Bar("")
bar.use_theme("macarons")
bar.add("月平均工资(元)", attr, v1, is_stack=True,is_label_show=True)
bar.render()
bar
attr = ["光明新区", "南山区", "宝安区", "盐田区","福田区", "罗湖区", "龙华新区", "龙岗区"]
v1 = aprice['count']
bar2 = Bar("深圳在招数据分析师岗位数量")
bar2.use_theme("infographic")
bar2.add("岗位数量", attr, v1, is_stack=True,is_label_show=True)
bar2.render()
bar2
```
