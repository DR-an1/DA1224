```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyecharts import Line
import os
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

plt.rcParams['figure.figsize'] = (8.0, 4.0)

os.chdir('C:/数据分析/美国婴儿')


#将1880到2010之间所有的婴儿姓名合并成一个大的Dataframe
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = './babynames/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)

    frame['year'] = year
    pieces.append(frame)

# Concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index=True)


total_births=names.pivot_table('births',index='year',columns='sex',aggfunc=sum)
total_births.tail(n=6)

#按性别和年度统计出生数
#total_births.plot(title='美国男女性增长速度', figsize=(10, 10), fontsize=20)
total_births.plot(kind='line')
plt.title(u'美国人口增长速度', fontsize=20)
plt.savefig("1.png")
plt.show()


#统计每一年各姓名婴儿的比例
def add_prop(group):
    group['prop']=group.births/group.births.sum()
    return group
names=names.groupby(['year','sex']).apply(add_prop)
names.groupby(['year','sex']).prop.sum()

#得到每对['year','sex']排名前1000的婴儿
def get_top1000(group):
    return group.sort_values(by='births',ascending=False)[:1000]
grouped=names.groupby(['year','sex'])
top1000=grouped.apply(get_top1000)
top1000.reset_index(inplace=True,drop=True)
boys=top1000[top1000.sex=='M']
girls=top1000[top1000.sex=='F']


#按year和name统计的总出生数透视表
total_births=top1000.pivot_table('births',index='year',columns='name',aggfunc=sum)


#绘制一些姓名的曲线图
subset=total_births[['John', 'Jordan', 'Mary', 'James']]
#subset.plot(subplots=True, figsize=(12, 10), grid=False,title="Number of births per year")
subset.plot(kind='line')
plt.title(u'热门姓名流行度透视', fontsize=20)
plt.savefig("2.png")
plt.show()

#得到名字总数排行榜
grouped=top1000.groupby(['name'])
popular_name=grouped['births']
aprice=popular_name.agg(['sum'])
aprice.sort_values('sum',ascending=False)[:20]


#计算最流行的1000个名字所占的比例
table = top1000.pivot_table('prop', index='year',columns='sex', aggfunc=sum)
#table.plot(title='Sum of table1000.prop by year and sex',yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2020, 10),figsize=(10, 8))
table.plot(kind='line')
plt.title(u'top1000的男女姓名和时间的关系', fontsize=20)
plt.savefig("3.png")
plt.show()

#计算占总出生人数(top1000)前50%的不同名字的数量
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
#diversity.plot(title="Number of popular names in top 50%",figsize=(10, 8))
diversity.plot(kind='line')
plt.title(u'前50%的不同名字的数量多样性分布', fontsize=20)
plt.savefig("4.png")
plt.show()


#各个年度名字长度的关系
top1000['n_count']=[len(n) for n in top1000['name']]
top1000.head(10)
grouped=top1000.groupby(['year', 'sex'])
def get_top(group):
    return group.sort_values(by='births',ascending=False)[:1]
top=grouped.apply(get_top)
total_births=top.pivot_table('n_count',index='year',columns='sex',aggfunc=sum)
#total_births.plot(title="Number of popular names in top count",figsize=(10, 8))
total_births.plot(kind='line')
plt.title(u'名字中字母个数与时间的关系', fontsize=20)
plt.savefig("5.png")
plt.show()


#计算各年度最流行的名字
def get_topname(group):
    return group.sort_values(by='births',ascending=False)[:1]
topname=grouped.apply(get_topname)
topname


#最后一个字母的变革
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'
table = names.pivot_table('births', index=last_letters,columns=['sex', 'year'], aggfunc=sum)
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
#计算各出生数占各年总人数的比例
letter_prop = subtable / subtable.sum()

#绘制各年度性别条形图
fig,axes=plt.subplots(2,1,figsize=(10,11))
letter_prop['M'].plot(kind='bar',rot=0,ax=axes[0],title='男性')
letter_prop['F'].plot(kind='bar',rot=0,ax=axes[1],title='女姓',legend=False)
plt.savefig("6.png")


letter_prop = table / table.sum()
dny_ts=letter_prop.loc[['d','n','y'],'M'].T
#dny_ts.plot(figsize=(10, 8))
dny_ts.plot(kind='line')
plt.title(u'姓名中末字母为d,n,y的变化图', fontsize=20)
plt.savefig("7.png")
plt.show()

#男孩名字与女孩名字的趋势
all_names = pd.Series(top1000.name.unique())
lesley_like = all_names[all_names.str.lower().str.contains('lesl')]
filtered=top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
table = filtered.pivot_table('births', index='year',columns='sex', aggfunc='sum')
table = table.div(table.sum(1), axis=0)
#绘制一张分性别的年度曲线图
#table.plot(style={'M': 'k-', 'F': 'k--'},figsize=(10, 8))
table.plot(kind='line',style={'M': 'k-', 'F': 'k--'})
plt.title(u'lesl开头的名字变化情况', fontsize=20)
plt.savefig("8.png")
plt.show()
```
