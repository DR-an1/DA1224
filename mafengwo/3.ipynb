```python
import pandas as pd
from pyecharts import Bar,Page, Grid
import os

os.chdir('C:/爬虫/蚂蜂窝')
page = Page()
data=pd.read_excel('C:\爬虫\蚂蜂窝\标签表.xlsx')

rankf=data.sort_values('food',ascending=False)[0:10]
bar=Bar('美食类城市排名',title_pos="left", width=800, height=400)
bar.use_theme("macarons")
attr=rankf['city']
v1=rankf['food']
bar.add("印象", attr, v1, is_stack=True,xaxis_rotate=30,is_label_show=True)
bar.render()

ranks=data.sort_values('sight',ascending=False)[0:10]
bar2=Bar('景点类城市排名',title_pos="left", width=800, height=400)
bar2.use_theme("infographic")
attr=ranks['city']
v1=ranks['sight']
bar2.add("印象", attr, v1, is_stack=True,xaxis_rotate=30,is_label_show=True,label_color='#FG4F94')
bar2.render()
'''
rankh=data.sort_values('happy',ascending=False)[0:10]
bar3=Bar('娱乐类城市排名',title_pos="left", width=800, height=400)
bar3.use_theme("vintage")
attr=rankh['city']
v1=rankh['happy']
bar3.add("印象", attr, v1, is_stack=True,xaxis_rotate=30,is_label_show=True,label_color='#000')
bar3.render()
'''
bar
```
