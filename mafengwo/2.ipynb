```python
#各类TOP
import pandas as pd
from pyecharts import Geo,Bar
import os

os.chdir('C:/爬虫/蚂蜂窝')

data=pd.read_excel('C:\爬虫\蚂蜂窝\游记表c.xlsx')
rank=data.sort_values('sum',ascending=False)[0:10]
bar=Bar('TOP10游记数量',title_pos="left", width=1000, height=400)
attr=rank['city']
v1=rank['sum']
bar.add("数量", attr, v1, is_stack=True,xaxis_rotate=30,is_label_show=True)
bar.render()
bar
data=pd.read_excel('C:\爬虫\蚂蜂窝\景点人气.xlsx')
rank=data.sort_values('number',ascending=False)[0:10]
bar=Bar('景点人气',title_pos="left", width=1000, height=600)
attr=rank['sight']
v1=rank['number']
bar.add("人气", attr, v1, is_stack=True,xaxis_rotate=29,is_label_show=True)
bar.render('1.html')
bar
data=pd.read_excel('C:\爬虫\蚂蜂窝\美食排名.xlsx')
rank=data.sort_values('number',ascending=False)[0:10]
bar=Bar('美食排名',title_pos="left", width=1000, height=600)
attr=rank['name']
v1=rank['number']
bar.add("人气", attr, v1, is_stack=True,xaxis_rotate=30,is_label_show=True)
bar.render()

bar

```
